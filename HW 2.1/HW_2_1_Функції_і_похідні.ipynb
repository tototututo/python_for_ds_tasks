### –î–æ–º–∞—à–Ω—ñ –∑–∞–≤–¥–∞–Ω–Ω—è –Ω–∞ –≤—ñ–¥–ø—Ä–∞—Ü—é–≤–∞–Ω–Ω—è –ø–æ–Ω—è—Ç—Ç—è —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó

1. –ó–∞–¥–∞–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –æ—Ü—ñ–Ω–∫–∏ –≤—ñ–¥ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω, –≤–∏—Ç—Ä–∞—á–µ–Ω–∏—Ö –Ω–∞ –Ω–∞–≤—á–∞–Ω–Ω—è: $ f(x) = 10x - 5 $.
  1. –ó –¥–æ–ø–æ–º–æ–≥–æ—é Python –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó.
  2. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –±—ñ–±–ª—ñ–æ—Ç–µ–∫—É `matplotlib` –¥–ª—è –ø–æ–±—É–¥–æ–≤–∏ –≥—Ä–∞—Ñ—ñ–∫—ñ–≤ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –∑–Ω–∞—á–µ–Ω—å –≤—ñ–¥ 0 –¥–æ 10 –≥–æ–¥–∏–Ω. –ü—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–π—Ç–µ, —è–∫ –∑–º—ñ–Ω–∞ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω –≤–∏–≤—á–µ–Ω–Ω—è –≤–ø–ª–∏–≤–∞—î –Ω–∞ –æ—Ü—ñ–Ω–∫—É.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')
```

```python
f = 10 * x - 5
```

```python
f_derivative = sp.diff(f, x)
print(f"–ü–æ—Ö—ñ–¥–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—ó f(x) = {f}:")
print(f"f'(x) = {f_derivative}")
```

```python
x_vals = np.linspace(0, 10, 100)
f_vals = 10 * x_vals - 5
f_deriv_vals = np.full_like(x_vals, 10)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label='f(x) = 10x - 5', color='blue')
plt.plot(x_vals, f_deriv_vals, label="f'(x) = 10", color='red', linestyle='--')

plt.title("–ì—Ä–∞—Ñ—ñ–∫ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó")
plt.xlabel("–ö—ñ–ª—å–∫—ñ—Å—Ç—å –≥–æ–¥–∏–Ω –Ω–∞–≤—á–∞–Ω–Ω—è (x)")
plt.ylabel("–û—Ü—ñ–Ω–∫–∞")
plt.legend()
plt.grid(True)
plt.show()
```

**–í–∏—Å–Ω–æ–≤–∫–∏**

- –§—É–Ω–∫—Ü—ñ—è `ùëì(ùë•) = 10ùë• ‚àí 5` –ø–æ–∫–∞–∑—É—î –ª—ñ–Ω—ñ–π–Ω—É –∑–∞–ª–µ–∂–Ω—ñ—Å—Ç—å: –∫–æ–∂–Ω–∞ –¥–æ–¥–∞—Ç–∫–æ–≤–∞ –≥–æ–¥–∏–Ω–∞ –Ω–∞–≤—á–∞–Ω–Ω—è –∑–±—ñ–ª—å—à—É—î –æ—Ü—ñ–Ω–∫—É –Ω–∞ 10 –±–∞–ª—ñ–≤.

- –ü–æ—Ö—ñ–¥–Ω–∞ `ùëì‚Ä≤(ùë•) = 10` –æ–∑–Ω–∞—á–∞—î, —â–æ
–∫–æ–∂–Ω–∞ –¥–æ–¥–∞—Ç–∫–æ–≤–∞ –≥–æ–¥–∏–Ω–∞ –Ω–∞–≤—á–∞–Ω–Ω—è –¥–æ–¥–∞—î —Ä—ñ–≤–Ω–æ 10 –±–∞–ª—ñ–≤ –¥–æ –æ—Ü—ñ–Ω–∫–∏.

2. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ Python —ñ `SymPy`, –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—ñ –¥–ª—è —Ñ—É–Ω–∫—Ü—ñ–π $ f(x) = \sin(x) $ —Ç–∞ $ f(x) = \ln(x) $. –ó–æ–±—Ä–∞–∑—ñ—Ç—å —Ü—ñ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—Ö–Ω—ñ –ø–æ—Ö—ñ–¥–Ω—ñ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é `matplotlib`.



```python
f1 = sp.sin(x)
f2 = sp.ln(x)
```

```python
f1_prime = sp.diff(f1, x)
f2_prime = sp.diff(f2, x)

print(f"–ü–æ—Ö—ñ–¥–Ω–∞ sin(x): {f1_prime}")
print(f"–ü–æ—Ö—ñ–¥–Ω–∞ ln(x): {f2_prime}")
```

```python
f1_lambda = sp.lambdify(x, f1, 'numpy')
f1p_lambda = sp.lambdify(x, f1_prime, 'numpy')

f2_lambda = sp.lambdify(x, f2, 'numpy')
f2p_lambda = sp.lambdify(x, f2_prime, 'numpy')
```

```python
x1_vals = np.linspace(-2 * np.pi, 2 * np.pi, 300)
x2_vals = np.linspace(0.1, 10, 300)
```

```python
f1_vals = f1_lambda(x1_vals)
f1p_vals = f1p_lambda(x1_vals)

f2_vals = f2_lambda(x2_vals)
f2p_vals = f2p_lambda(x2_vals)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x1_vals, f1_vals, label='f(x) = sin(x)', color='blue')
plt.plot(x1_vals, f1p_vals, label="f'(x) = cos(x)", color='red', linestyle='--')
plt.title("–ì—Ä–∞—Ñ—ñ–∫ sin(x) —Ç–∞ –π–æ–≥–æ –ø–æ—Ö—ñ–¥–Ω–æ—ó")
plt.xlabel("x")
plt.ylabel("–ó–Ω–∞—á–µ–Ω–Ω—è")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x2_vals, f2_vals, label='f(x) = ln(x)', color='green')
plt.plot(x2_vals, f2p_vals, label="f'(x) = 1/x", color='orange', linestyle='--')
plt.title("–ì—Ä–∞—Ñ—ñ–∫ ln(x) —Ç–∞ –π–æ–≥–æ –ø–æ—Ö—ñ–¥–Ω–æ—ó")
plt.xlabel("x")
plt.ylabel("–ó–Ω–∞—á–µ–Ω–Ω—è")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```


3. –ó–∞–≥–∞–ª—å–Ω—ñ –≤–∏—Ç—Ä–∞—Ç–∏ –∫–æ–º–ø–∞–Ω—ñ—ó –≤—ñ–¥ –æ–±—Å—è–≥—É –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ –∑–∞–¥–∞–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—î—é $ f(x) = x^2 - 5x + 1 $, –¥–µ $ x $ ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏—Ä–æ–±–ª–µ–Ω–∏—Ö –æ–¥–∏–Ω–∏—Ü—å. –û–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó –∑ –¥–æ–ø–æ–º–æ–≥–æ—é SymPy —Ç–∞ –≤—ñ–∑—É–∞–ª—ñ–∑—É–π—Ç–µ –∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó —ñ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –Ω–∞ –≤—ñ–¥—Ä—ñ–∑–∫—É –≤—ñ–¥ 0 –¥–æ 100.

```python
f = x**2 - 5*x + 1
```

```python
f_prime = sp.diff(f, x)
print(f"–§—É–Ω–∫—Ü—ñ—è –≤–∏—Ç—Ä–∞—Ç: f(x) = {f}")
print(f"–ü–æ—Ö—ñ–¥–Ω–∞: f'(x) = {f_prime}")
```

```python
f_lambda = sp.lambdify(x, f, 'numpy')
f_prime_lambda = sp.lambdify(x, f_prime, 'numpy')
```

```python
x_vals = np.linspace(0, 100, 500)
f_vals = f_lambda(x_vals)
f_prime_vals = f_prime_lambda(x_vals)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(x_vals, f_vals, label='f(x) = x¬≤ - 5x + 1', color='blue')
plt.plot(x_vals, f_prime_vals, label="f'(x) = 2x - 5", color='red', linestyle='--')

plt.title("–ó–∞–≥–∞–ª—å–Ω—ñ –≤–∏—Ç—Ä–∞—Ç–∏ —Ç–∞ —ó—Ö –∑–º—ñ–Ω–∞ –∑–∞–ª–µ–∂–Ω–æ –≤—ñ–¥ –æ–±—Å—è–≥—É –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞")
plt.xlabel("–ö—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏—Ä–æ–±–ª–µ–Ω–∏—Ö –æ–¥–∏–Ω–∏—Ü—å (x)")
plt.ylabel("–í–∏—Ç—Ä–∞—Ç–∏")
plt.legend()
plt.grid(True)
plt.show()
```
