1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')
```

```python
f = 10 * x - 5
```

```python
f_derivative = sp.diff(f, x)
print(f"Похідна функції f(x) = {f}:")
print(f"f'(x) = {f_derivative}")
```

```python
x_vals = np.linspace(0, 10, 100)
f_vals = 10 * x_vals - 5
f_deriv_vals = np.full_like(x_vals, 10)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label='f(x) = 10x - 5', color='blue')
plt.plot(x_vals, f_deriv_vals, label="f'(x) = 10", color='red', linestyle='--')

plt.title("Графік функції та її похідної")
plt.xlabel("Кількість годин навчання (x)")
plt.ylabel("Оцінка")
plt.legend()
plt.grid(True)
plt.show()
```

**Висновки**

- Функція `𝑓(𝑥) = 10𝑥 − 5` показує лінійну залежність: кожна додаткова година навчання збільшує оцінку на 10 балів.

- Похідна `𝑓′(𝑥) = 10` означає, що
кожна додаткова година навчання додає рівно 10 балів до оцінки.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
f1 = sp.sin(x)
f2 = sp.ln(x)
```

```python
f1_prime = sp.diff(f1, x)
f2_prime = sp.diff(f2, x)

print(f"Похідна sin(x): {f1_prime}")
print(f"Похідна ln(x): {f2_prime}")
```

```python
f1_lambda = sp.lambdify(x, f1, 'numpy')
f1p_lambda = sp.lambdify(x, f1_prime, 'numpy')

f2_lambda = sp.lambdify(x, f2, 'numpy')
f2p_lambda = sp.lambdify(x, f2_prime, 'numpy')
```

```python
x1_vals = np.linspace(-2 * np.pi, 2 * np.pi, 300)
x2_vals = np.linspace(0.1, 10, 300)
```

```python
f1_vals = f1_lambda(x1_vals)
f1p_vals = f1p_lambda(x1_vals)

f2_vals = f2_lambda(x2_vals)
f2p_vals = f2p_lambda(x2_vals)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x1_vals, f1_vals, label='f(x) = sin(x)', color='blue')
plt.plot(x1_vals, f1p_vals, label="f'(x) = cos(x)", color='red', linestyle='--')
plt.title("Графік sin(x) та його похідної")
plt.xlabel("x")
plt.ylabel("Значення")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x2_vals, f2_vals, label='f(x) = ln(x)', color='green')
plt.plot(x2_vals, f2p_vals, label="f'(x) = 1/x", color='orange', linestyle='--')
plt.title("Графік ln(x) та його похідної")
plt.xlabel("x")
plt.ylabel("Значення")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
f = x**2 - 5*x + 1
```

```python
f_prime = sp.diff(f, x)
print(f"Функція витрат: f(x) = {f}")
print(f"Похідна: f'(x) = {f_prime}")
```

```python
f_lambda = sp.lambdify(x, f, 'numpy')
f_prime_lambda = sp.lambdify(x, f_prime, 'numpy')
```

```python
x_vals = np.linspace(0, 100, 500)
f_vals = f_lambda(x_vals)
f_prime_vals = f_prime_lambda(x_vals)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(x_vals, f_vals, label='f(x) = x² - 5x + 1', color='blue')
plt.plot(x_vals, f_prime_vals, label="f'(x) = 2x - 5", color='red', linestyle='--')

plt.title("Загальні витрати та їх зміна залежно від обсягу виробництва")
plt.xlabel("Кількість вироблених одиниць (x)")
plt.ylabel("Витрати")
plt.legend()
plt.grid(True)
plt.show()
```
